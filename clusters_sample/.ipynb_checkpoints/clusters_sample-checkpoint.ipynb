```python
from PIL import Image,ImageFilter,ImageDraw
import numpy as np
import pandas as pd
import pickle

from matplotlib import pyplot as plt
from matplotlib import cm as CM
```

```python
test_path = '../TRANCOS_v3/images'
```

**Cluster# 14:**<br>
Highest positive errors since it combines high values for all the metrics <br>
(brightness which is specially high, groundtruth, overlapping and dispersion) <br>


```python
sample_df14 = pd.read_pickle('./sample_df14.pkl')
sample_df14
```

```python
fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(5):
    ax = fig.add_subplot(4, 5, i+1)
    ax.imshow(Image.open(test_path+"/"+sample_df14['im_path'][i][-18:]))
    
for i in range(5):
    ax = fig.add_subplot(4, 5, i+6)
    ax.axis('off')
    ax.text(0.3, 0.99, sample_df14['im_path'][i][-7:], fontsize=13)
    ax.text(0, 0.7, "Density maps:", fontsize=12)
    ax.text(0.1, 0.5, "- Ground_Truth", fontsize=12)
    ax.text(0.1, 0.3, "- Estimation", fontsize=12)
       
for i in range(5):
    ax = fig.add_subplot(5, 5, i+11)
    ax.imshow(sample_df14['dens_im'][i],cmap=CM.jet)
    
for i in range(5):
    ax = fig.add_subplot(5, 5, i+16)
    ax.imshow(sample_df14['resImg'][i],cmap=CM.jet) 

```
